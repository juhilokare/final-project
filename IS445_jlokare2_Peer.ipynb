```python
import pandas as pd
import ipywidgets 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
data = pd.read_csv("/Users/juhilokare/Desktop/freedom.csv", encoding='unicode_escape', sep=';')
```

```python
data
```

```python
countries = ['Albania', 'Austria', 'Belarus', 'Belgium', 'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 'Portugal', 'Romania', 'Serbia', 'Slovak Republic', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine', 'United Kingdom' ]
year = 2021
```

```python
filtered_data = data[(data['Countries'].isin(countries)) & (data['Year'] == year)].copy()
```

```python
filtered_data = filtered_data.sort_values('5Di  Market openness', ascending=True)
filtered_data = filtered_data.sort_values('Economic Freedom Summary Index', ascending=True)
```

```python
#This scatter plot shows the correlation between economic freedom and market openness across European countries in 2021. 
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(filtered_data['Economic Freedom Summary Index'], filtered_data['5Di  Market openness'], alpha=0.5)
plt.title('Scatter Plot of Economic Freedom vs. Market Openness')
plt.xlabel('Economic Freedom Summary Index')
plt.ylabel('Market Openness')
plt.grid(True)
plt.show()
```

```python
filtered_data = filtered_data.sort_values('2A  Judicial independence', ascending=True)
filtered_data = filtered_data.sort_values('2B  Impartial courts', ascending=True)
```

```python
#This scatter plot shows the correlation between judicial independence and impartiallity of courts across European countries in 2021. 
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(filtered_data['2A  Judicial independence'], filtered_data['2B  Impartial courts'], alpha=0.5)
plt.title('Scatter Plot of Judicial independence vs. Impartial courts')
plt.xlabel('Judicial independence')
plt.ylabel('Impartial courts')
plt.grid(True)
plt.show()
```
