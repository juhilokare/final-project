```python
import pandas as pd
import ipywidgets 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
data = pd.read_csv("/Users/juhilokare/Desktop/freedom.csv", encoding='unicode_escape', sep=';')
```

```python
countries = ['Albania', 'Austria', 'Belarus', 'Belgium', 'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 'Portugal', 'Romania', 'Serbia', 'Slovak Republic', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine', 'United Kingdom' ]
year = 2021
```

```python
filtered_data = data[(data['Countries'].isin(countries)) & (data['Year'] == year)].copy() 
```

```python
filtered_data
```

```python
filtered_data.loc[:, '1A Government consumption'] = filtered_data['1A Government consumption'].str.replace(',', '.').astype(float)
```

```python
filtered_data = filtered_data.sort_values('1A Government consumption', ascending=True)
```

```python
#This bar chart shows government consumption by European countries in 2021. 
```

```python
plt.figure(figsize=(10, 8))
plt.bar(filtered_data['Countries'], filtered_data['1A Government consumption'], color='blue')
plt.xlabel('Countries')
plt.ylabel('Government Consumption')
plt.title('Government Consumption by European Countries')
plt.xticks(rotation=90)  
plt.show()
```

```python
filtered_data = filtered_data.sort_values('1C  Government investment', ascending=True)
```

```python
#This bar chart shows government investment by European countries in 2021. 
```

```python
plt.figure(figsize=(10, 8))
plt.bar(filtered_data['Countries'], filtered_data['1C  Government investment'], color='blue')
plt.xlabel('Countries')
plt.ylabel('Government Investment')
plt.title('Government Investment by European Countries')
plt.xticks(rotation=90)
plt.show()
```
